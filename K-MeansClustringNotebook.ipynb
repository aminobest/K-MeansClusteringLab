```python
#import the used libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import seaborn as sns
```

```python
# Import data
df = pd.read_csv("input/data.csv")
```

```python
#The imported data has the following attributes

#Developer ID: An anoymized ID assigned to each developer
#Error Frequency: Number of Coding Errors / (Number of Coding Errors + Number of Successful Compilations)
#Shortcut Usage: Number of Times Shortcuts Used / Total Number of Keypresses in the IDE
#Usage of Help Features: Number of Times Help Features Accessed / Total Number of Interactions with the IDE
#Breakpoint Usage: Number of Breakpoints Set / Total Number of Debugging Actions
```

```python
##########################
#
# Clustering, visualizing the clusters and reporting insights 
#
# Task 2. a. Try out clustering with different K values and features, then visualize the results. 
#         b. Report any emerging patterns observed in the data when trying different k values and features 
#
#         In the following, we are providing clustering examples with visualizations
#
##########################
```

```python
# Example of clustering (1)
```

```python
# KMeans clustering
k = 2 #Input the number of clusters
features_set = ['Error Frequency', 'Shortcut Usage'] #Input the feature set from ['Error Frequency', 'Shortcut Usage', 'Help Feature Usage', 'Breakpoint Usage']

kmeans = KMeans(n_clusters=k, random_state=0).fit(df[features_set])

# Make a copy of df, Add the cluster labels to the copy
clustered_df = df.copy(deep=True)
clustered_df['Cluster'] = kmeans.labels_
```

```python
# Plotting the clusters 2D in space
feature1 = 'Error Frequency' #Input the feature to plot 
feature2 = 'Shortcut Usage' #Input the feature to plot
plt.figure(figsize=(10,7))
sns.scatterplot(data=clustered_df, x=feature1, y=feature2, hue='Cluster', palette='viridis', s=100, alpha=0.7)
plt.title('Clusters of Developers')
plt.show()
```

```python
# Example of clustering (2)
```

```python
# KMeans clustering
k = 3 #Input the number of clusters
features_set = ['Error Frequency', 'Breakpoint Usage'] #Input the feature set from ['Error Frequency', 'Shortcut Usage', 'Help Feature Usage', 'Breakpoint Usage']

kmeans = KMeans(n_clusters=k, random_state=0).fit(df[features_set])

# Make a copy of df, Add the cluster labels to the copy
clustered_df = df.copy(deep=True)
clustered_df['Cluster'] = kmeans.labels_
```

```python
# Plotting the clusters in 2D space
feature1 = 'Error Frequency' #Input the feature to plot 
feature2 = 'Breakpoint Usage' #Input the feature to plot
plt.figure(figsize=(10,7))
sns.scatterplot(data=clustered_df, x=feature1, y=feature2, hue='Cluster', palette='viridis', s=100, alpha=0.7)
plt.title('Clusters of Developers')
plt.show()
```
